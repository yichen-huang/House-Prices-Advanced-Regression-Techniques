```python
import pandas as pd
import numpy as np
#implementing the new column 'PriceType

data = pd.read_csv('train.csv', sep=',')

data['PriceType'] = np.where(
                            data['SalePrice'] >= 300000,
                            'HIGH',
                            np.where(
                                data['SalePrice'] <= 150000,
                                'LOW',
                                'MEDIUM'
                                )
                            )
data

data_2 = data.iloc[:, 21:39].copy()
data_2 = pd.concat([data_2, data.iloc[:, -2:]], axis=1)

data_2.dropna(subset=['MasVnrType'], inplace=True)  # Remove rows with missing values in the specified column
data_2.dropna(subset=['MasVnrArea'], inplace=True)
data_2




```

```python
#creating the function to comparee the pricetype with the categorical columns

import matplotlib.pyplot as plt

def data_show_categorical(attribute, title=None):

    
    # previous plots
    #plt.rcParams.update({'font.size': 5})
    #data['SalePrice'].hist(by=data[attribute])

    data[attribute].value_counts().plot(kind = 'bar')
    plt.show()
    data.groupby([attribute, 'PriceType']).size().unstack().plot(kind='bar', stacked=True, title=title)

def data_show_categorical_compare(attribute, attribute2, title=None):
    data.groupby([attribute, 'PriceType', attribute2]).size().unstack().plot(kind='bar', stacked=True, title=title)
```

```python
from scipy.stats import gaussian_kde
import numpy as np
import seaborn as sns


def data_show_numerical(attribute):
    # first plot
    data.plot(kind = 'scatter', x = attribute, y = 'SalePrice')
    plt.show()

    # second plot
    tmp_data = data.dropna(subset=[attribute])

    xy = np.vstack([tmp_data[attribute], tmp_data['SalePrice']])
    z = gaussian_kde(xy)(xy)

    fig, ax = plt.subplots()
    ax.scatter(tmp_data[attribute], tmp_data['SalePrice'], c=z, s=10)
    plt.show()

    # third plot
    data[attribute].plot.kde()

"""def data_show_numerical2(attribute):
    sns.catplot(x=attribute, y='PriceType', data=data, palette='Spectral')
    plt.show()"""

def data_show_numerical2(attribute, title=None):
    top_15_values = data[attribute].value_counts().nlargest(15).index
    filtered_data = data[data[attribute].isin(top_15_values)]
    sns.catplot(x=attribute, y='PriceType', data=filtered_data, palette='Spectral')
    plt.show()
#only the 15 biggest values, otherwise, we can't have a good legend

def data_show_numerical_compare(attribute, attribute2, title=None):
    # sns.catplot(data=data, x=attribute, y="SalePrice", kind="swarm", title=title)
    sns.lmplot(x=attribute, y='SalePrice', data=data, hue=attribute2, fit_reg=False, palette='flare')
    plt.show()

"""def data_show_numerical_compare_legend(attribute, attribute2, title=None):
    # sns.catplot(data=data, x=attribute, y="SalePrice", kind="swarm", title=title)
    sns.catplot(x=attribute, y='PriceType', data=data, hue=attribute2, palette='dark:#5A9_r')
    
    # Get the unique values of attribute2
    unique_values = data[attribute2].unique()
    
    # Select a subset of values from unique_values for the legend
    num_values = min(len(unique_values), 5)
    selected_values = unique_values[:num_values]
    
    # Create a custom legend with the selected values
    plt.legend(selected_values)
    
    # Show the plot
    plt.show()"""

def data_show_numerical_compare2(attribute, attribute2, title=None):
    # sns.catplot(data=data, x=attribute, y="SalePrice", kind="swarm", title=title)
    sns.catplot(x=attribute, y='PriceType', data=data, hue=attribute2, palette='dark:#5A9_r')
    plt.show()

def data_show_numerical_compare3(attribute, attribute2, title=None):
    # sns.catplot(data=data, x=attribute, y="SalePrice", kind="swarm", title=title)
    sns.catplot(x=attribute, y='PriceType', data=data, hue=attribute2, palette='ch:s=.25,rot=-.25', kind='box')
    plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Create a copy of the data DataFrame
data_encoded = data_2.copy()

# Iterate through each column
for column in data_encoded.columns:
    # Check if the column is categorical (object or string type)
    if data_encoded[column].dtype == 'object' or isinstance(data_encoded[column].dtype, list):
        # Create an instance of LabelEncoder
        label_encoder = LabelEncoder()
        # Encode the column to integer codes
        #if it is a numerical column, it don't encode it
        data_encoded[column] = label_encoder.fit_transform(data_encoded[column])

# Calculate the correlation matrix for all columns
correlation_matrix = data_encoded.corr()

# List to store correlation pairs
categorical_pairs = []
numerical_pairs = []

# Iterate through each column pair
for col1 in correlation_matrix.columns:
    for col2 in correlation_matrix.columns:
        # Skip if the columns are the same or if either column is 'SalePrice'
        if col1 == col2 or col1 == 'SalePrice' or col2 == 'SalePrice' or col1 == col2 or col1 == 'PriceType' or col2 == 'PriceType':
            continue
        
        # Calculate the correlation between the columns
        correlation = correlation_matrix.loc[col1, col2]
        
        # Check the data types of the columns
        if col1 in ['RoofStyle', 'RoofMatl', 'Exterior1st', 'Exterior2nd', 'MasVnrType', 'ExterQual', 'ExterCond', 'Foundation', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2']:
            categorical_pairs.append((col1, col2, correlation))
        elif col1 in ['MasVnrArea', 'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF'] and col2 in ['MasVnrArea', 'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF']:
            numerical_pairs.append((col1, col2, correlation))

#remove the duplicates of categorical columns
for pair1 in categorical_pairs:
    for pair2 in categorical_pairs:
        if pair1[0] == pair2[1] and pair1[1] == pair2[0]:
            categorical_pairs.remove(pair2)

# Sort the categorical pairs by correlation value
categorical_pairs.sort(key=lambda pair: pair[2], reverse=True)

# Select the top 5 categorical pairs
top_5_categorical_pairs = categorical_pairs[:5]

# Print the top 5 categorical pairs
print("Top 5 Categorical Pairs:")
for pair in top_5_categorical_pairs:
    print(pair)

print()

#remove the duplicates of numerical columns
for pair1 in numerical_pairs:
    for pair2 in numerical_pairs:
        if pair1[0] == pair2[1] and pair1[1] == pair2[0]:
            numerical_pairs.remove(pair2)

# Sort the categorical pairs by correlation value
numerical_pairs.sort(key=lambda pair: pair[2], reverse=True)

# Select the top 5 categorical pairs
top_5_numerical_pairs = numerical_pairs[:5]

# Print the top 5 categorical pairs
print("Top 5 Numerical Pairs:")
for pair in top_5_numerical_pairs:
    print(pair)
```
