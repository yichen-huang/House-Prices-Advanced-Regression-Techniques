```python
# Handle missing values - Example 2: Removal of rows with missing values
selected_columns.dropna(subset=['MasVnrType'], inplace=True)  # Remove rows with missing values in the specified column
selected_columns.dropna(subset=['MasVnrArea'], inplace=True)

# for these attributes, this is not missing values : BsmtQual, BsmtCond, BsmtExposure, BsmtFinType1, BsmtFinType2.
selected_columns.shape[0] #the new number of rows in selected_columns
```

```python
import pandas as pd

train = pd.read_csv('train.csv', sep=',')

saleprice = train['SalePrice'] #add the column SalePrice
select = pd.concat([selected_columns, saleprice], axis=1)
select
```

```python
correlations = select.corr()['SalePrice'].drop('SalePrice')
print(correlations)
```

```python
#now correlations with all the columns compared to the sale price

non_integer_columns = select.select_dtypes(exclude=['int']).columns
#encoded_select = pd.get_dummies(select, columns=['non_integer_columns'])
encoded_select = pd.get_dummies(select, columns=non_integer_columns)
correlations_2 = encoded_select.corr()['SalePrice']

# Sort correlations in descending order
sorted_correlations = correlations_2.sort_values(ascending=False)

# Select the top 10 strongest positive correlations
top_10_positive = sorted_correlations.head(10)

# Select the top 10 strongest negative correlations
top_10_negative = sorted_correlations.tail(10)

# Combine the top positive and negative correlations
top_correlations = pd.concat([top_10_positive, top_10_negative])

print(top_correlations)
```

```python
final_correlation = pd.concat([top_correlations, correlations]).sort_values(ascending=False)
print(final_correlation)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Remove the top correlation
final_correlation = final_correlation.iloc[2:]

f, ax = plt.subplots(figsize=(12, 9))

sns.heatmap(final_correlation.to_frame(), annot=True, cmap='coolwarm', square=True)

plt.suptitle("Correlatation Feature HeatMap with SalePrice")
plt.xlabel("Features")
plt.ylabel("Features")
```
