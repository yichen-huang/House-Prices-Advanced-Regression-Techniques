# GarageFinish

```python
# Catégories de GarageFinish
categories = ['RFn', 'Unf', 'Fin']

# Compter les occurrences de LOW, MEDIUM et HIGH pour chaque catégorie de GarageFinish
counts_low = [TrainData[TrainData['GarageFinish'] == category]['PriceType'].value_counts().get('LOW', 0) for category in categories]
counts_medium = [TrainData[TrainData['GarageFinish'] == category]['PriceType'].value_counts().get('MEDIUM', 0) for category in categories]
counts_high = [TrainData[TrainData['GarageFinish'] == category]['PriceType'].value_counts().get('HIGH', 0) for category in categories]

# Créer le graphique à barres groupées
bar_width = 0.25
index = np.arange(len(categories))

plt.bar(index, counts_low, width=bar_width, label='LOW')
plt.bar(index + bar_width, counts_medium, width=bar_width, label='MEDIUM')
plt.bar(index + 2*bar_width, counts_high, width=bar_width, label='HIGH')

plt.xlabel('Catégorie')
plt.ylabel('Occurrences')
plt.title('Occurrences des catégories dans les catégories de GarageFinish')
plt.xticks(index + bar_width, categories)
plt.legend()

plt.show()
```

Coefficient de corrélation

```python
# Convertir les catégories de GarageFinish en valeurs numériques
mapping = {'RFn': 0, 'Unf': 1, 'Fin': 2}
TrainData['GarageFinish_Num'] = TrainData['GarageFinish'].map(mapping)

# Calculer le coefficient de corrélation entre GarageFinish_Num et le prix de la maison
correlation = TrainData['GarageFinish_Num'].corr(TrainData['SalePrice'])
print("Coefficient de corrélation : ", correlation)

if correlation > 0.3 or correlation < -0.3:
    print("Il y a corrélation")
else:
    print("Il n'y a pas de corrélation")
```

```python
# Matrice de corrélation
correlation_matrix = np.corrcoef([TrainData['OpenPorchSF'], TrainData['EnclosedPorch'], TrainData['3SsnPorch']])

# Seuil de corrélation
threshold = 0.3

# Variables corrélées
correlated_vars = []

# Parcourir la matrice de corrélation et trouver les variables corrélées
for i in range(len(correlation_matrix)):
    for j in range(i+1, len(correlation_matrix)):
        correlation = correlation_matrix[i, j]
        if abs(correlation) > threshold:
            correlated_vars.append((i, j))

# Afficher les variables corrélées ou "Pas de corrélation"
if len(correlated_vars) > 0:
    for var1, var2 in correlated_vars:
        print(f"Les variables {var1} et {var2} sont corrélées.")
else:
    print("Pas de corrélation significative entre les variables.")
```
