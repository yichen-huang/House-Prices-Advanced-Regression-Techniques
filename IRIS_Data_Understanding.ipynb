```python
iris['y'].value_counts().plot(kind = 'bar', color = ['red', 'blue', 'green'])
plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize = (15, 7))
iris['sepal length'].hist(ax = ax[0, 0], legend=True)
iris['sepal width'].hist(ax = ax[0, 1], legend=True)
iris['petal length'].hist(ax = ax[1, 0], legend=True)
iris['petal width'].hist(ax = ax[1, 1], legend=True)
plt.show()
```

```python
iris.plot(kind = 'scatter', x = 'sepal length', y = 'sepal width')
plt.show()
```

Choosing Colormaps in Matplotlib: https://matplotlib.org/stable/tutorials/colors/colormaps.html

```python
encCol = iris['y'].replace({'Iris-setosa': 'yellow', 'Iris-versicolor': 'red', 'Iris-virginica': 'blue'})
iris.plot(kind = 'scatter', x = 'sepal length', y = 'sepal width', c = encCol, cmap = 'plasma')
plt.show()
```

```python
encCol = iris['y'].replace({'Iris-setosa': 1, 'Iris-versicolor': 2, 'Iris-virginica': 3})
iris.plot(kind = 'scatter', x = 'sepal length', y = 'sepal width', c = encCol, cmap = 'plasma')
plt.show()
```

```python
import seaborn as sn
```

```python
sn.pairplot(iris, hue = 'class')
plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize = (15, 7))
sn.boxplot(y = 'sepal length', x = 'class', data = iris, ax = ax[0,0])
sn.boxplot(y = 'sepal width', x = 'class', data = iris, ax = ax[0,1])
sn.boxplot(y = 'petal length', x = 'class', data = iris, ax = ax[1,0])
sn.boxplot(y = 'petal width', x = 'class', data = iris, ax = ax[1,1])
plt.show()
```
